### Initial inspection and visualization

```python
stocks.info()
stocks.describe()
close=stocks.loc[:,'Close'].copy()#its the closing-price of stocks that matters.
```

```python
close.plot(figsize=(12,8),fontsize=11)#stocks starting as different level.
#for comparing performance, base-level has to be same for all, can be done with normalization.
```

```python
#we can get the normalised base-level by dividing prior first level from whole
#following changes after, with normalized value of 100,
close.iloc[0]
norm=close.div(close.iloc[0]).mul(100)
norm
```

```python
norm.plot(figsize=(12,8))#worst performance is IBM after normalizing.
```

```python
###shift() method for calculating absolute/relative change of yesterday to today.
close
aapl=close.AAPL.copy().to_frame()#creating a df for apple for example to play with.
aapl
```

```python
#to see price increase or decrease from last trading day.
#shift(), shifts the value by desired value.
aapl['lag1'] = aapl.shift(periods=1)#no prior value,NaN.periods=2,-1. Generally makes sense in +ve.
```

```python
aapl['diff']=aapl.AAPL.sub(aapl.lag1)#daily increase of decrease.
aapl
```

```python
#for finance, it makes sense to track relative increase,decrease
aapl['pct_change']=aapl.AAPL.div(aapl.lag1).sub(1).mul(100)#sub(1) for change in percentage terms.
```

```python
aapl
```

```python
#shift() method is general.we can do above with diff() and pct_change() method. 
#diff of one trading day to another.
aapl.AAPL.diff(periods=1)#same as above.
aapl.AAPL.pct_change(periods=1).mul(100)#same as above.
```

```python
#above for daily changes, but for monthly changes,with monthly frequency,but on business days,
aapl.AAPL.resample('BM').last().pct_change(periods=1).mul(100)
# we can do same for quaterly or weekly as well.
#we get monthly returns with pct_change().
```

```python
### Performance based on Risk and Return.
norm.plot()
```

```python
#mean return and std return
ret=close.pct_change().dropna()#daily return
```

```python
summary=ret.describe().T.loc[:,['mean','std']] #for mean and std of return
summary#small numbers,
```

```python
#to annualize,
summary['mean']= summary['mean']*252 #avg cal year has 252 Business Days. 
summary['std']= summary['std']*np.sqrt(252)
```

```python
summary#annualized changes.
```

```python
#higher risk and return with x=std and y=mean?
#IBM was the worst.
summary.plot(kind='scatter',x='mean',y='std',figsize=(15,12),s=50,fontsize = 15)
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,'std'],summary.loc[i,'mean']),size=15)
plt.title('Risk/Return',fontsize = 15)
plt.xlabel('ann. Risk(std)',fontsize = 15)
plt.ylabel('ann. Return',fontsize = 20)
plt.show()#I dont know how upper and lower are different.
```

```python
summary.plot(kind = "scatter", x = "std", y = "mean", figsize = (15,12), s = 50, fontsize = 15)
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, "std"], summary.loc[i, "mean"]), size = 15)
plt.xlabel("ann. Risk(std)", fontsize = 15)
plt.ylabel("ann. Return", fontsize = 15)
plt.title("Risk/Return", fontsize = 20)
plt.show()
```

### Handling Missing Values with bfill,ffill,interpolation.

```python
close#contains stock prices on business days.
all_days=pd.date_range('2009-12-31','2019-02-06',freq='D')
all_days
```

```python
close.reindex(all_days)#here, we can use method='ffill','bfill' etc like fillna().
```

```python
#for missing values, first we can fill NaN values for month_end, weekday and day_name.
close.month_end = close.index.is_month_end
close.weekday= close.index.weekday
close.day_name=close.index.day_name()
```

```python
#for missing values,
close.fillna(method='ffill',inplace=True)
#we can fill scaler values, doesnt make sense here.can be 'fill'
```
