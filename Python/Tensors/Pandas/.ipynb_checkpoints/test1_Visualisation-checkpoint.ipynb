```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn')#available styles.
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.plot(subplots=True,figsize= (12,12),sharex=False, sharey=False)
plt.show() # default as line plot.
```

```python
titanic.age.plot(figsize=(12,8),fontsize=15, c='r',
                 linestyle='-',xlim =(0,900),ylim=(0,90),xticks= xticks,yticks=yticks,rot=45)
plt.title('Titanic - Ages', fontsize = 15)
plt.legend(loc = 'best',fontsize=15)
plt.xlabel('Passenger No', fontsize = 15)
plt.ylabel('Age',fontsize=15)
plt.grid()
plt.show()
```

```python
xticks = [x for x in range(0,900,50)]
xticks
```

```python
yticks = [y for y in range(0,90,5)]
yticks
```

```python
titanic.age.value_counts(bins=10)
```

### Histogram

```python
titanic.age.plot(kind='hist',figsize=(12,8), fontsize=15,bins=80)
plt.show()
```

```python
titanic.age.hist(figsize=(12,8),bins=80,xlabelsize=15,ylabelsize=15)
plt.show()
```

```python
#direct matplotlib hist method, which can not handle missing values.
plt.figure(figsize=(12,8))
plt.hist(titanic.age.dropna(),bins=80, density=True, cumulative= True)
plt.show()#density for relative quantity, cumulative contains all of data.
```

### Barchart and Piechart for aggregated data

```python
summer_2012 = pd.read_csv('summer_2012.csv',index_col = 'Country')
```

```python
summer_2012.Medal.plot(kind='barh',figsize=(12,8),fontsize=15)
plt.show()
```

```python
summer_2012.Medal.plot(kind='pie',figsize=(12,8),fontsize=15)
plt.show()
```

### Scatterplots

```python
titanic.plot(kind='scatter',figsize=(12,8),fontsize=15, 
             x= 'age', y= 'fare', c='r',marker = 'x', s= 20)
plt.show()
#can only be used for df. Requires x and y column.
```

```python
#to convert 2-D scatter plot into 3-D plot, pass another column in c.
titanic.plot(kind='scatter',figsize=(12,8),fontsize=15, 
             x= 'age', y= 'fare', c='survived',marker = 'x', s= 20, colormap='viridis')
plt.show()
```
