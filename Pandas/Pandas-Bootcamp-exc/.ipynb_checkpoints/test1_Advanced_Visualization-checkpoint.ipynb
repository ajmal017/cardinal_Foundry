```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic.csv')
```

### First Steps

```python
plt.figure(figsize=(12,8))
sns.countplot(data=titanic,x='sex',hue='pclass')
#count categories with bars. y='sex' horizontal. Cant pass both x and y together I suppose.
#subgroups with hue by passing categorical data.
sns.set(font_scale=2,palette='viridis')
plt.show()#sex as categorical data with even pclass with groupby helps.countplot same with visual.
```

### Categorical plots.

```python
#swarmplot is same as stripplot(scatter plot with one categorical var) but without jitter.
#swarmplot for seeing better distribution.
#jitter true for non over-lapping dots.
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.stripplot(data=titanic,x='sex',y='age',jitter=True,hue='pclass',dodge=True)
#dodge for seperate category-wise visual.
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.swarmplot(data=titanic,x='sex',y='age',hue='pclass',dodge=True)#non overlapping freq distbn.
```

```python
#violinplot visualizes frequency distribution even better.Curve is kde.
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.swarmplot(data=titanic,x='sex',y='age',hue='pclass',dodge=True,color='black')
sns.violinplot(data=titanic,x='sex',y='age',jitter=True,hue='pclass',dodge=True)
#we can combine violinplot with swarmplot.color parameter different that disticts one from other.
```

```python
#if two hue levels for hue parameter, then they can be combined together with split.
sns.violinplot(data=titanic,x='pclass',y='age',jitter=True,hue='sex',dodge=True,split=True)
```

```python
sns.barplot(data=titanic,x='pclass',y='age',hue='sex',dodge=True)
#barplot calculates mean age for each group. Point estimates because mean over groups.
#black line gives confidence interval, verticvally for 95% cases.
```

```python
sns.pointplot(data=titanic,x='pclass',y='age',hue='sex',dodge=True)
#overlapping of confidence interval for 2nd class.
#Males are older than females with statistical significance for 1st and 3rd class.
```

### Joint/Regression Plot.

```python
#similar to scatter plot. two numerical continuous var here. Provides frequency distribution.
#doesnt work as good with categorical data.
#shaded area represents 95% confidence interval and here is positive corelation,can be +ve or -ve.
#kind=kde, hex, scatter available.
sns.set(font_scale=1.5)
sns.jointplot(data=titanic,x='age',y='fare',kind='reg',height=8)
```

```python
#similar to reg plot, lmplot.
sns.lmplot(data=titanic,x='age',y='fare',aspect=1,hue='sex')
#subplots with col='sex'and row='sex'
```

```python
#logistic reg because survived col is categorical with 0 and 1.
sns.lmplot(data=titanic,x='age',y='survived',aspect=1,col='sex',logistic=True)
```

### Matrixplots/Heatmaps

```python
#for cross tabular data with categorical data.
pd.crosstab(titanic.sex,titanic.pclass)
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.2)
sns.heatmap(pd.crosstab(titanic.sex,titanic.pclass),annot=True,cmap='Reds',vmax=150)
#annot for annoting exact numbers.cmap for color.vmax for capping like in outliers.
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(pd.crosstab(titanic.sex,titanic.pclass, values= titanic.survived,aggfunc='mean'),
            annot= True,cmap='Reds')
```

```python
sns.heatmap(titanic.corr(),cmap='Reds',annot=True)
```
