### pd.date_range() for creating a customized datetime with range.


```python
#incremental date range
pd.date_range(start='1-1-2015',end='2-2-2015')#default freq='D'.'D' defines to daily.
```

```python
pd.date_range(end='2-2-2015',freq= 'D',periods=32)#includes end.
pd.date_range(start='2-2-2015',freq= 'D',periods=32)#includes start.
#We have to define three parameters.Not four.
```

```python
pd.date_range(end='2-2-2015',freq= 'B',periods=32)#'B' defines to business days.
```

```python
pd.date_range(end='2-2-2015',freq= 'H',periods=32)#Hourly data.
```

```python
pd.date_range(start='2-2-2015',freq= 'W',periods=32)#default 'W-SUN'.From 2nd, 32 sundays.
pd.date_range(start='2-2-2015',freq= 'W-WED',periods=32)#Wed here.
```

```python
pd.date_range(start='2-2-2015',freq= 'M',periods=32)#last day of every month.
pd.date_range(start='2-2-2015',freq= 'MS',periods=32)#first day.
pd.date_range(start='2-2-2015',freq= pd.DateOffset(months=1),periods=32)#starts with 2-2-2015.
```

```python
pd.date_range(start='2-2-2015',freq= 'Q',periods=32)#Quaterly.Last day of quater.
pd.date_range(start='2-2-2015',freq= 'QS',periods=32)#first day of quater.default 'QS-JAN'
pd.date_range(start='2-2-2015',freq= 'QS-May',periods=32)#quater starts from May insead APR.
```

```python
pd.date_range(start='2-2-2015',freq= 'A',periods=32)#Annually.Last day of year.
pd.date_range(start='2-2-2015',freq= 'AS',periods=32)#First day of year.
pd.date_range(start='2-2-2015',freq= 'AS-Nov',periods=32)#Annually from NOV.
pd.date_range(start='2-2-2015',freq= pd.DateOffset(years=1),periods=32)#from 2-2-15 Annually.
```

```python
#increased flexibility.
pd.date_range(start='2-2-2015',freq= '3H2D',periods=32)#increament of 3 hours and 2 days.
```

```python
#downsampling - converting from high to low frequency D,W,H,with resample(),similar to groupby().
temp
temp.info()
temp.resample('D')#DatetimeIndexResampler 
```

```python
list(temp.resample('D'))#doesnt work somehow.Hence no point for len(list(temp.resample('D')))
```

```python
temp.resample('D').first()#first timestamp by chaining a aggfunc on resampler.
```

```python
temp.resample('D').last()
temp.resample('D').mean()
temp.resample('D').sum()#doesnt make sense here.
temp.resample('D').median()
temp.resample('2H').first()
temp.resample('W').mean()
temp.resample('W-WED').mean()
temp.resample('M').mean()
temp.resample('MS').mean()
temp.resample('MS',loffset='14D').mean()#adjust resample time lables.
temp.resample('Q').mean()
temp.resample('Q-Feb').mean()
temp.resample('Y').mean()
temp.resample('YS').mean()
```

```python
temp.resample('M',kind='period').mean()#kind=timestamp is default.
#'MS' gives error because we show whole month in period.No need for start or end.
temp.resample('W',kind='period').mean()
temp.resample('Q',kind='period').mean()
temp.resample('A',kind='period').mean()
```

```python
temp_m = temp.resample('M',kind='period').mean()
```

```python
temp_m.info()
temp.index[0]
```

```python
temp_m.plot(figsize=(12,8))#with downsapling we smoothened the data.
```

```python
temp.plot(figsize=(12,8))#many daily oscillations with many data points.
```

#### Pitfalls of period index.

```python
temp_mts =temp.resample('M',kind='timestamp').mean()
```

```python
#behaviour for datetime index.
temp_mts.index
temp_mts.loc['2015-2']
temp_mts.loc['2015-2':'2015-11']
```

```python
#behaviour of period index.
temp_m.info()
temp_m.loc['2015']#same as datetime index.
temp_m.loc['2015-2':'2015-11']#same as datetime index.
temp_m.loc['2015']#not all rows in 2015, like in datetime index but just one.
```

```python
#to convert period index into timestamp index,
temp_m.to_timestamp(how='start')#first timestamps. With how='end', we get last.
```

```python
temp_mts.info()
```

```python
birth_old= pd.date_range(start='26-11-2015',periods=4,freq=pd.DateOffset(years=1))
```

```python
temp.loc[birth_old]#we get temp when its available.
#this raises keyerror. Use .reindex() instead. Lets explore why.
```

```python
#reindex() allows optional filling logic.
temp.reindex(birth_old,method='bfill')#we still get NaN.But we can take care of it.
temp.reindex(birth_old,method='ffill')
temp.reindex(birth_old,method='nearest')
#but sometimes, such as in this case, its better to leave out missing values.
```
