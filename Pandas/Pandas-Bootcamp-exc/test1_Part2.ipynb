### Importing Data

```python
pd.read_csv('titanic.csv', index_col= 'class', header= 0,
                      names=['alive','class','gender','age', 'sibsp', 'parch','price','emb','deck'])
# by default, header=0 i.e first row becomes the header for df.
```

```python
titanic =pd.read_csv('titanic.csv', index_col='pclass', usecols=['survived', 'pclass', 'sex', 'age'])
#usecols for choosing only specific columns for analysis.
```

```python
titanic.columns =['alive', 'gender','age']
```

```python
titanic.index.name = 'class'
```
