```python
titanic.groupby(['sex']).apply(group_mean)#mean with func to df.
```

```python
titanic.nlargest(5,'age')
```

```python
def five_oldest_surv(group):#slighly more complex user defined func
    return group[group.survived==1].nlargest(5,'age')#filter first, then largest method to group.
```

```python
titanic.groupby(['sex']).apply(five_oldest_surv)
```

### Stack and Unstack for groupby

```python
medals_per_country = summer.groupby(['Country','Medal']).Medal.count()
```

```python
medals_per_country
```

```python
medals_per_country.loc[('USA','Gold')]
```

Unstack

```python
medals_per_country.unstack(level =-2,fill_value=0)#changes shape based on level.
#by default, -1 inner index, which becomes the rightmost index to be unstack,creating 3 columns.
```

```python
medals_per_country=medals_per_country.unstack(level =-1,fill_value=0)
```

```python
#to change the order of columns.
medals_per_country= medals_per_country[['Gold','Silver','Bronze']]
```

```python
medals_per_country.sort_values(by=['Gold','Silver','Bronze'], 
                               ascending = [False,False,False],inplace=True)
```

```python
medals_per_country
```

```python
#many countries, plot looks horrible. Choose 10 countries.
medals_per_country.head(10).plot(kind='bar',figsize=(12,8),fontsize=15)
plt.show()
plt.style.use('seaborn')
```

```python
#stack method is opposite of unstack method.
#where unstack method turns inner index into columns, stack converts to multi-index again.
medals_per_country.stack()
```

```python
medals_per_country.stack().unstack()
```
