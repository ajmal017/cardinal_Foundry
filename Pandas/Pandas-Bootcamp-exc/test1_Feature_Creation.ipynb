### Discretization and binning

```python
age_bins = [0,10,18,30,55,100]
pd.cut(titanic.age,age_bins,right=False)#rightmost inclusive or exclusive
#dtype categoyry in order.
```

```python
group_names=['child','teenager','young_adult','adult','elderly']
titanic['age_category']=pd.cut(titanic.age, age_bins,right=False, labels=group_names)
#converting numerical bins to labels.
```

```python
titanic.age_category.value_counts()
```

```python
titanic.groupby('age_category').survived.mean()
#no linear relationship here.So,convert to categorical with labels.
```

```python
titanic['fare_category']=pd.cut(titanic.fare,5,precision=2)#approx divided in 20%
```

```python
titanic.fare_category.value_counts()#5 equal width cuts dont make sense here.Let's customize it.
```

```python
#qcut is quantile cut, with respect to passengers.
titanic['fare_category']=pd.qcut(titanic.fare,5, precision=2)
```

```python
fare_labels = ['very_cheap','cheap','moderate','exp','very_exp']
titanic['fare_category']=pd.qcut(titanic.fare,[0,0.1,0.25,0.50,0.90,1],precision=2,
                                 labels=fare_labels)
```

```python
titanic.fare_category.value_counts()
```

```python
titanic.groupby(['age_category','fare_category']).survived.mean().unstack()
```

### Scaling/Standardization

```python
#age and fare have different range/scale of values. One higher, other lower.
titanic.fare.plot(figsize=(12,8))
titanic.age.plot(figsize=(12,8))
plt.show()
```

```python
#with z-score, we can standardize these values.z= z-a/std
mean_age= titanic.age.mean()
mean_fare=titanic.fare.mean()
std_age=titanic.age.std()
std_fare=titanic.fare.std()
titanic['age_z']= round((titanic.age-mean_age)/std_age,2)
titanic['fare_z']= round((titanic.fare-mean_fare)/std_fare,2)
```

```python
titanic.head()
```

```python
titanic.age_z.plot()
titanic.fare_z.plot()#now both values seem to be in similar range/scale.
```

### String Operations

```python
summer=pd.read_csv('summer.csv')
summer.Athlete=summer.Athlete.str.title()#can't assign at the end of sentense a.k.a func call.
```

```python
summer[['Surname','First_Name']]=summer.Athlete.str.split(', ',n=1,expand=True)#n is first appearance here.
#expand becomes true for two separate columns.
```

```python
#to remove white space
summer['Surname']=summer.Surname.str.strip()
summer['First_Name']= summer.First_Name.str.strip()
```

```python
summer.drop(columns='Athlete',inplace=True)
```
